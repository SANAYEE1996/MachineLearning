```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.common.keys import Keys
import time
from tqdm.notebook import tqdm
```

```python
header = "https://www.youtube.com/channel/"
data = pd.read_csv("C:/Users/dudtk/0YSP/CrawlingPractice/sample/뉴스정치개수.csv", sep=",", encoding='UTF8')
primList = data['주키']
MonthCount_list = []
count = len(primList)
```

```python
for i in range(0,1):
    geturl = header + primList[i]
    driver = webdriver.Chrome(executable_path='C:/Users/dudtk/0YSP/chromedriver')
    driver.get(url=geturl)
    time.sleep(1)
    menulist = driver.find_elements_by_tag_name('tp-yt-paper-tab > div')
    try:
        menulist[1].click()
        title = driver.find_elements_by_tag_name('div > div > ytd-channel-name > div > div > yt-formatted-string')[0].text
        time.sleep(1)
        MonthCount = 0
        while(True):
                cntList = driver.find_elements_by_tag_name('ytd-grid-video-renderer > div > div > div > div > div > div > span')
                isExit = False
                MonthCount = 0
                for j in range(len(cntList):
                    day = cntList[j].text
                    lastDay = cntList[len(cntList)-1].text
                    if '시간 전' in day:
                        MonthCount += 1
                    elif '일 전' in day:
                        MonthCount += 1
                    elif '주 전' in day:
                        MonthCount += 1
                    elif '개월' in day:
                        isExit = True
                        break
                    elif '년' in day:
                        isExit = True
                        break
                    elif '시간 전' in lastDay or '일 전' in lastDay or '주 전' in lastDay :
                        break
                if(isExit):
                    break
                time.sleep(1.0)
                driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    except:
        print((i+1)," 번째 순위 안됌.")
        MonthCount = 0
    if i < count :
        data.loc[i,'한달 업로드 영상 수'] = MonthCount
    time.sleep(2)
    driver.quit()
```
